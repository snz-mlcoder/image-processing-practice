```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab import files

# Upload image
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# Read image in grayscale
gray_image = cv2.imread(filename, 0)

# Initialize the FAST detector
fast = cv2.FastFeatureDetector_create()

# Detect keypoints with non-max suppression
keypoints_with_nonmax = fast.detect(gray_image, None)
print("Number of keypoints with non-max suppression:", len(keypoints_with_nonmax))

# Draw keypoints on the image
img_with_nonmax = cv2.drawKeypoints(gray_image, keypoints_with_nonmax, None, color=(0,255,0))

# Display the image using matplotlib
plt.figure(figsize=(8, 8))
plt.imshow(img_with_nonmax, cmap='gray')
plt.title('FAST keypoints - with non-max suppression')
plt.axis('off')
plt.show()

# Disable non-max suppression
fast.setNonmaxSuppression(False)

# Detect keypoints again
keypoints_without_nonmax = fast.detect(gray_image, None)
print("Number of keypoints without non-max suppression:", len(keypoints_without_nonmax))

# Draw keypoints on the image
img_without_nonmax = cv2.drawKeypoints(gray_image, keypoints_without_nonmax, None, color=(0,255,0))

# Display the image using matplotlib
plt.figure(figsize=(8, 8))
plt.imshow(img_without_nonmax, cmap='gray')
plt.title('FAST keypoints - without non-max suppression')
plt.axis('off')
plt.show()
```
