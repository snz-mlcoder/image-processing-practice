```python

import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # Required for displaying images in Colab
from google.colab import files

# Upload an image (user selects file)
uploaded = files.upload()

# Get the name of the uploaded file (assumes one file)
file_name = next(iter(uploaded))

# Read the image
img = cv2.imread(file_name)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = np.float32(gray)

# Apply Harris Corner Detection
dst = cv2.cornerHarris(gray, blockSize=2, ksize=21, k=0.04)

# Optional: dilate the result to enhance corner points
dst = cv2.dilate(dst, None)

# Mark detected corners in red on the original image
img[dst > 0.01 * dst.max()] = [0, 0, 255]

# Display the final image
cv2_imshow(img)
```
