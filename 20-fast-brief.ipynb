```python
# Install OpenCV with contrib (for BRIEF)
!pip install opencv-contrib-python

import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab import files

# Upload image
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# Read the image in grayscale
gray_image = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)

# Initialize FAST detector
fast = cv2.FastFeatureDetector_create()

# Detect keypoints using FAST
keypoints = fast.detect(gray_image, None)

# Initialize BRIEF descriptor
brief = cv2.xfeatures2d.BriefDescriptorExtractor_create()

# Compute descriptors with BRIEF
keypoints, descriptors = brief.compute(gray_image, keypoints)

# Draw keypoints
img_keypoints = cv2.drawKeypoints(gray_image, keypoints, None, color=(0,0,200))

# Show result
plt.figure(figsize=(8, 8))
plt.imshow(img_keypoints, cmap='gray')
plt.title('BRIEF Keypoints (with FAST)')
plt.axis('off')
plt.show()

# Optional: Print descriptor shape
print("Number of keypoints:", len(keypoints))
print("Descriptor shape:", descriptors.shape if descriptors is not None else "No descriptors")
```
